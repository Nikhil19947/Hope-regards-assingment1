```python
from tkinter import *
import os

tasks = []
tasks = ["Note1","Note2","Note3","Note4","Note5","Note5","Note6","Note7","Note8","Note9","Note10"]


def delete2():
    screen3.destroy()
    
def delete3():
    screen4.destroy()
    
def delete4():
    screen5.destroy()
    
def saved():
    screen8 = Toplevel(screen)
    screen8.title("Saved")
    Label(screen8, text = "Saved").pack()
    
    
    
def save():
    filename = raw_filename.get()
    notes = raw_notes.get()
    
    data = open(filename, "w")
    data.write(notes)
    data.close()
    
    saved()
    
def create():
    global raw_filename
    raw_filename=StringVar()
    global raw_notes
    raw_notes=StringVar()
    
    screen7 = Toplevel(screen)
    screen7.title("Info")
    Label(screen7,text="Please enter a filename  for the note below:").pack()
    Entry(screen7,textvariable=raw_filename).pack()
    Label(screen7,text="Please enter Notes for file below:").pack()
    Entry(screen7,textvariable=raw_notes).pack()
    Button(screen7,text="Save",command=save).pack()
    
def veiw_notes():
    filename1 = raw_filename1.get()
    data = open(filename1,"r")
    data1 = data.read()
    screen10 = Toplevel(screen)
    screen10.title("Notes")
    Label(screen10,text=data1).pack()
    
    
    
def veiw():
    screen9 = Toplevel(screen)
    screen9.title("Info")
    Label(screen9,text="Welcome to dashboard").pack()
    all_files = os.listdir()
    Label(screen9, text="Please use one of the name below").pack()
    Label(screen9, text= all_files).pack()
    global raw_filename1
    
    raw_filename1 = StringVar()
    
    Entry(screen9, textvariable=raw_filename1).pack()
    Button(screen9, text = "OK",command=veiw_notes).pack()


def delete_note1():
    filename3 = raw_filename2.get()
    os.remove(filename3)
    screen12 = Toplevel(screen)
    screen12.title("Notes")
    Label(screen12,text=filename3+"removed").pack()
    
  

    
def delete_note():
    screen11 = Toplevel(screen)
    screen11.title("Info")
    Label(screen11,text="Welcome to dashboard").pack()
    all_files = os.listdir()
    Label(screen11, text="Please use one of the name below").pack()
    Label(screen11, text= all_files).pack()
    global raw_filename2
    
    raw_filename2 = StringVar()
    
    Entry(screen11, textvariable=raw_filename2).pack()
    Button(screen11, text = "OK",command=delete_note1).pack()
    
def edit_note2():
    screen8 = Toplevel(screen)
    screen8.title("Edited")
    Label(screen8, text = "Saved").pack()
    
     
    
def edit_note1():
    filename = raw_filename.get()
    notes = raw_notes.get()
    
    data = open(filename, "w")
    data.write(notes)
    data.close()
    
    edit_note2()
    
    
def edit():
    global raw_filename
    raw_filename=StringVar()
    global raw_notes
    raw_notes=StringVar()
    
    screen13 = Toplevel(screen)
    screen13.title("Info")
    Label(screen13,text="Please enter a filename  for the Note you want to edit:").pack()
    Entry(screen13,textvariable=raw_filename).pack()
    Label(screen13,text="Please enter new format:").pack()
    Entry(screen13,textvariable=raw_notes).pack()
    Button(screen13,text="Save",command=edit_note1).pack()
   
    
    
    
def session():
    screen6 = Toplevel(screen)
    screen6.title("dashboard")
    Label(screen6,text="Welcome to dashboard").pack()
    Button(screen6,text="Creat Note",command=create).pack()
    Button(screen6,text="Veiw Note",command=veiw).pack()
    Button(screen6,text="Edit Note",command=edit).pack()
    Button(screen6,text="Delete Note",command=delete_note).pack()
    

def login_sucess():
    session()
    
    
def password_not_recognised():
    global screen4
    screen4 = Toplevel(screen)
    screen4.title("sucess")
    #screen4.geometry("1366*768")
    Label(screen4, text="Password Incorrect").pack()
    Button(screen4, text="OK",command = delete3).pack()
    

def user_not_found():
    global screen5
    screen5 = Toplevel(screen)
    screen5.title("sucess")
    #screen5.geometry("1366*768")
    Label(screen5, text="User Not Found Error").pack()
    Button(screen5, text="OK",command = delete4).pack()

    



def Register_User():
    print("working")
    UserName_info= UserName.get()
    Password_info= Password.get()
    
    file=open(UserName_info,"w")
    file.write(UserName_info+"\n")
    file.write(Password_info)
    file.close()
    
    UserName_Entry.delete(0,END)
    Password_Entry.delete(0,END)
    
    Label(screen1,text="Registration is suceessfull",fg="green",font=("Calibri",25)).pack()


    
def login_verify():
    UserName1=UserName_verify.get()
    Password1=Password_verify.get()

    UserName_Entry1.delete(0,END)
    Password_Entry1.delete(0,END)
    
    list_of_files=os.listdir()
    if UserName1 in list_of_files:
        file1=open(UserName1,"r")
        verify=file1.read().splitlines()  
        if Password1 in verify:
            login_sucess()
        else:
            password_not_recognised()
    else:
        user_not_found()

    
    
    
def Register():
    global screen1
    screen1=Toplevel(screen)
    screen1.title("Register")
    
    global UserName
    global Password
    
    global UserName_Entry 
    global Password_Entry
    
    UserName=StringVar()
    Password=StringVar()
    
    
    Label(screen1,text="Please enter details below ").pack()
    Label(screen1,text="").pack()
    
    Label(screen1,text="UserName *").pack()
    UserName_Entry= Entry(screen1,textvariable= UserName)
    UserName_Entry.pack()
    Label(screen1,text="Password *").pack()
    Password_Entry= Entry(screen1,textvariable= Password)
    Password_Entry.pack()
    Label(screen1,text="").pack()
    Button(screen1,text="Register",width=10,height=1,command=Register_User).pack()


def login():
    global screen2
    screen2 = Toplevel(screen)
    screen2.title("Login")
    Label(screen2,text="Please enter details below for login ").pack()
    Label(screen2,text="").pack()
    
    global UserName_verify
    global Password_verify
    
    global UserName_Entry1
    global Password_Entry1
    
    UserName_verify=StringVar()
    Password_verify=StringVar()
    
    Label(screen2,text="UserName *").pack()
    UserName_Entry1=Entry(screen2,textvariable= UserName_verify)
    UserName_Entry1.pack()
    Label(screen2,text="Password *").pack()
    Password_Entry1=Entry(screen2,textvariable= Password_verify)
    Password_Entry1.pack()

    Label(screen2,text="").pack()
    
    Button(screen2,text="Login",width=10,height=1,command=login_verify).pack()
    
def update_listbox():
    #clear_listbox()
    for task in tasks:
        lb_tasks.insert("end" ,task)
    
    
def add_task():
    task = txt_input.get()
    if task!="":
        tasks.append(task)
        update_listbox()


def note_funct():
    global screen14
    global txt_input
    global lb_tasks
    screen14 = Toplevel(screen)
    Button(screen14,text="Title",width=10,height=1).pack()
    txt_input=Entry(screen14,width=25)
    txt_input.pack()
    Button(screen14,text="ADD",width=10,height=1,command=add_task).pack()
    Label(screen14,text="Discription",width=10,height=1).pack()
    lb_tasks=Listbox(screen14,width=100,height=15)
    lb_tasks.pack()
    
    
def main_screen():
    global screen
    screen=Tk()
    #screen.geometry("300*250")
    screen.title("To-Do-List App")
    Button(text="Note",bg="pink",height="1",width="15",font=("Calibri",25),command=note_funct).pack()
    Label(text="").pack()
    #Label(text="Note 1",bg="blue", font=("Calibri",15)).pack()
    #Label(text="").pack()
    Button(text="Login",bg="pink",height="2",width="30",font=("Calibri",25),command=login).pack()
    Label(text="").pack()
    Button(text="Register",bg="pink",height="2",width="30",font=("Calibri",25),command=Register).pack()
        
    screen.mainloop()
        
main_screen()
          
```
